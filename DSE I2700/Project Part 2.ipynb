```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt, mpld3
import matplotlib.dates as mdates
import numpy as np
import datetime
from bokeh.plotting import figure, output_file, show
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
%matplotlib inline
```

```python
df = pd.read_csv("project_data.csv", index_col="DATE", parse_dates=["DATE"])
```

```python
df['week']=df.index.week
```

```python
dow = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

plt.figure(figsize=(18,12))
for week, sp in zip(range(23,40), range(1,17)):
    weekly = df[df["week"] == week]   
    plt.subplot(4,4,sp)
    plt.plot(weekly.index, 'merissa_steps', data=weekly, label='Merissa')
    plt.plot(weekly.index, 'sheridan_steps', data=weekly, label='Sheridan')
    plt.plot(weekly.index, 'tova_steps', data=weekly, label='Tova')
    plt.ylabel('Steps')
    plt.xlabel('Day')
    plt.legend()
    plt.xticks(weekly.index, dow)
    plt.title('Week {}'.format(week))
plt.tight_layout()
plt.savefig('weeklystepviz.png')
```
