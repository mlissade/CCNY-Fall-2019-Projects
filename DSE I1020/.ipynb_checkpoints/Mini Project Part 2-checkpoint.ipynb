For this mini project, I used the heart disease data set from the UC Irvine Machine Learning Repository. I chose this data set because it was used for a research that used machine learning on patient data to predict the presence of heart disease or not. Due to my family's medical history of heart disease, this type of research always piques my interest. From this data set, the hope is to learn if we can predict the presence of heart disease in a patient based off of the patient's medical information. I believe that by using this data we can apply a machine learning method to make this prediction.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'processed_cleveland.data'
col_names = ['age','sex','cp','trestbps','chol','fbs','restecg',
             'thalach','exang','oldpeak','slope','ca','thal','num']
df = pd.read_csv(filename, delimiter=',', header=None, names=col_names, na_values='?').dropna()
```

When I inspected the data the first time around, I noticed that any missing values were denoted by `'?'` so I had any of those entries with missing data dropped.

```python
df.num.replace([2, 3, 4], 1, inplace=True)
df['diagnosis'] = np.where(df['num'] > 0, 'yes', 'no')

num_vars = ['age','trestbps','chol','thalach','oldpeak']
categ_vars = ['sex','cp','fbs','restecg','exang','slope','thal','ca', 'num']
target = df['diagnosis']
```

Experiments with this partiular database have concentrated on simply attempting to distinguish presence of heart disease (values 1,2,3,4 of the variable `num`) from absence (value 0). For the purpose of my project, I decided to simplify this variable by creating the variable, `diagnosis`, to turn the `num` variable from a value between 0 and 4 to just indicate `'yes'` disease or `'no'` disease.

## Initial EDA

```python
df.info()
df.describe()
```

#### Variables
1. `age`: in years
2. `sex`: (1 = male; 0 = female)
3. `cp`: chest pain type (1=typical angina; 2=atypical angina; 3=non-ginal pain; 4=asymptomatic)
4. `trestbps`: resting blood pressure (in mmHg on admission to hospital)
5. `chol`: serum cholestoral in mg/dl
6. `fbs`: fast blood sugar > 120 mg/dl (1=true; 0=false)
7. `restecg`: resting electrocardiographic results (0=normal; 1=having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV); 2=showing probable or definite left ventricular hypertrophy by Estes' criteria)
8. `thalach`: maximum heart rate achieved
9. `exang`: exercise induced angina (1 = yes; 0 = no)
10. `oldpeak`: ST depression induced by exercise relative to rest
11. `slope`: the slope of the peak exercise ST segment (1=upsloping; 2=flat; 3=downsloping)
12. `ca`: number of major vessels (0-3) colored by flourosopy
13. `thal`: 3 = normal; 6 = fixed defect; 7 = reversable defect
14. `num`: (the predicted attribute):  diagnosis of heart disease (angiographic disease status) (0 = < 50% diameter narrowing; 1 = >= 50% diameter narrowing)
15. `diagnosis`: I created this variable to turn the `num` variable from a value between 0 and 4 to just indicate `yes` disease or `no` disease

## EDA: Looking at the correlations between the data
In this part, I decided to re-do some of my previous visualizations however this time applying my target variable, `diagnosis`, as another dimension to see if there is anything noteworthy.

```python
sns.pairplot(df, vars=num_vars, hue='diagnosis', palette='husl')
```

Along the diagonals, each variable's distribution of patients with heart disease versus patients without seem to have a lot of overlap. The two variables that I feel like have the best separation are `age` and `thalach`. It surprises me that that there is no large difference with the `chol`. These might be the variables that could work best for my target variable. I would assume that the distribution of patients diagnosed with heart disease would show that those patients have higher cholestoral levels.

```python
plt.figure(figsize=(15,10))
for x, sp in zip(categ_vars, range(1,10)):
    plt.subplot(3,3,sp)
    sns.countplot(x=df[x], hue=df.diagnosis, palette='husl')   
plt.tight_layout()
```

I did not feel like adding the third dimension in this case helped give any insight.

```python
plt.figure(figsize=(20,10))
for y, sp in zip(num_vars, range(1,6)):
    plt.subplot(1,5,sp)
    sns.violinplot(y=df[y], x=df.diagnosis, hue=df.sex, palette='husl', split=True)   
plt.tight_layout()
```

For this visualization, I created violin plots for my numerical variables with `diagnosis` on the x-axis.
